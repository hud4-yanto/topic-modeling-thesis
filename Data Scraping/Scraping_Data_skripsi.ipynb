```python
link_list = []
tahun_list = []
kampus_list = []

#Cari skripsi SI di rentang tahun yang telah ditentukanz
for thn in range(int(Tahun_awal),int(Tahun_akhir)):
    resp = request.urlopen("http://digilib.uinsa.ac.id/view/divisions/fst=5Fsi/"+ str(thn)+".html").read()
    soup = BeautifulSoup(resp, "html")
    skripsi = soup.find_all('div', class_='ep_view_page ep_view_page_view_divisions')


    for list in skripsi:
        #Ambil hrefnya
        for link in list.find_all('a'):
            for links in link:
                href = link.get('href')
                if href and href.startswith('http://digilib.uinsa.ac.id/'):
                    # Ambil ID skripsinya
                    href_rapi = href.split('/')
                    link_list.append(href_rapi[3])


        #Ambil tahun
        for tahun in list.find_all('p'):
            tahun_string = tahun.get_text()

            # buang spasi yang tidak diperlukan
            tahun_rapi = tahun_string.split("\n")
            tahun_fix = tahun_rapi[4].replace(')','')
            tahun_fixx = tahun_fix.replace('(','')

            # append judul ke list
            tahun_list.append(tahun_fixx)

            #ambil kampus
            kampus_list.append('UIN Surabaya')
    print(resp[1:1000])
```
