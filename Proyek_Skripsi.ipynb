# Modeling

Inisiation Reduce Dimension DAta, CLuster data, vectorizer and representation data

```python
from umap import UMAP

umap_model2 = UMAP(n_neighbors=25, n_components=10, min_dist=0.5, metric='cosine', random_state=42)
```

```python
from hdbscan import HDBSCAN

hdbscan_model24 = HDBSCAN(min_cluster_size=20, metric='euclidean', cluster_selection_method='eom', prediction_data=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model24 = CountVectorizer(stop_words=kalimat_stop,max_df=100, min_df=1, max_features=1000, ngram_range=(1, 1))
```

Topic Modeling start

```python
from bertopic import BERTopic
topic_model24 = BERTopic(

  # Pipeline models
  embedding_model=embedding_model1,
  umap_model=umap_model2,
  hdbscan_model=hdbscan_model24,
  vectorizer_model=vectorizer_model24,

  # representation_model=representation_model,

  # Hyperparameters
  top_n_words=20,
  verbose=True
)

topics, probs = topic_model24.fit_transform(abstrak_fix, embeddings1)
```

## Evaluate model

Topic Coherence

```python
# Extract vectorizer and tokenizer from BERTopic
vectorizer = topic_model24.vectorizer_model
tokenizer = vectorizer.build_tokenizer()

# Extract features for Topic Coherence evaluation
words = vectorizer.get_feature_names_out()
tokens = [tokenizer(doc) for doc in abstrak_fix]
dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(token) for token in tokens]
topic_words = [[words for words, _ in topic_model24.get_topic(topic)]
               for topic in range(len(set(topics))-1)]

# Evaluate
coherence_model = CoherenceModel(topics=topic_words,
                                 texts=tokens,
                                 corpus=corpus,
                                 dictionary=dictionary,
                                 coherence='c_v')
coherence = coherence_model.get_coherence()
print("Topic coherence IndoSBERT:", coherence)
```

Topic diversity

```python
from octis.evaluation_metrics.diversity_metrics import TopicDiversity

topic_words = {'topics':[[words for words, _ in topic_model24.get_topic(topic)]
               for topic in range(len(set(topics))-1)]}

metric = TopicDiversity() # Initialize metric
topic_diversity_score = metric.score(topic_words) # Compute score of the metric
print(topic_diversity_score)
```
